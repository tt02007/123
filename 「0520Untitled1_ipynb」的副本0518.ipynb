```python
#字串加密器:)
#拿來罵人的首選

import string
import random
try:
    strin = input("請輸入要加密的字串(限英文) ")
    length_of_string = int(input('填的字串長度 '))

except:
    print("格式不符")


k = 0
stri = ''
for i in range(len(strin)):
    stri += strin[k]
    stri += ''.join(random.choice(string.ascii_letters) for _ in range(length_of_string))
    k += 1

stri += str(length_of_string+1)
print('密碼: ',stri)
```
